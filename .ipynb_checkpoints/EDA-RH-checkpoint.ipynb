```python
import pandas as pd
import matplotlib.pyplot as plt

def plotarDiferenca(df, coluna, grupo_col, grupo1_val, grupo2_val, grupo1_label='Grupo 1', grupo2_label='Grupo 2'):
    """
    Plota um gráfico de barras comparando a média de uma coluna específica entre dois grupos.

    df: DataFrame contendo os dados
    coluna: Coluna a ser comparada
    grupo_col: Coluna que define os grupos
    grupo1_val: Valor que identifica o primeiro grupo
    grupo2_val: Valor que identifica o segundo grupo
    grupo1_label: Rótulo do primeiro grupo
    grupo2_label: Rótulo do segundo grupo
    """
    # Calculando as médias para cada grupo
    media_grupo1 = df[df[grupo_col] == grupo1_val][coluna].mean()
    media_grupo2 = df[df[grupo_col] == grupo2_val][coluna].mean()

    # Criando o DataFrame para plotar
    data_comparacao = {
        'Grupo': [grupo1_label, grupo2_label],
        'Média': [media_grupo1, media_grupo2]
    }

    df_comparacao = pd.DataFrame(data_comparacao)

    # Plotando o gráfico de barras
    plt.figure(figsize=(12, 8))
    plt.bar(df_comparacao['Grupo'], df_comparacao['Média'], color=['red', 'blue'])

    # Adicionando rótulos e título
    plt.xlabel('Grupo')
    plt.ylabel('Média')
    plt.title(f'Comparação de Médias da Coluna "{coluna}"')
    
    # Adicionando anotações
    for i in range(len(df_comparacao)):
        plt.text(i, df_comparacao['Média'][i] + 0.1, round(df_comparacao['Média'][i], 2), ha='center')

    # Exibindo o gráfico
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
dfModelo = df.loc[(df['Nivel_Cargo'] == 'Sênior') | (df['Nivel_Cargo'] == 'Pleno')]
```

```python
dfModelo = dfModelo[["Pontuacao_Desempenho","Satisfacao_Trabalho","Nivel_Cargo","Salario","Horas_Extras","Desligamento"]]
```

```python
dfModelo
```

```python
dfModelo['Nivel_Cargo'].isnull().sum()
```

```python
dfModelo['Nivel_Cargo'].unique()
```

```python
one_hot_encoder = OneHotEncoder(sparse=False)
nivel_cargo_encoded = one_hot_encoder.fit_transform(dfModelo[['Nivel_Cargo']])
nivel_cargo_encoded
```

```python
nivel_cargo_encoded_df = pd.DataFrame(nivel_cargo_encoded, columns=one_hot_encoder.get_feature_names_out(['Nivel_Cargo']))
nivel_cargo_encoded_df
```

```python
dfModelo
```

```python
dfModelo = pd.concat([dfModelo.drop('Nivel_Cargo', axis=1).reset_index(drop=True), nivel_cargo_encoded_df], axis=1)
dfModelo
```

```python
dfModelo[["Nivel_Cargo_Pleno","Nivel_Cargo_Sênior","Desligamento"]].sum()
```

```python
# Dividindo os dados em X e y
X = dfModelo.drop('Desligamento', axis=1)
y = dfModelo['Desligamento']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=485)
```

```python
y_test.count()
```

```python
y_train.count()
```

```python
# Treinando o modelo de árvore de decisão
tree_model = DecisionTreeClassifier(random_state=30)
tree_model.fit(X_train, y_train)

# Fazendo previsões com a árvore de decisão
tree_predictions = tree_model.predict(X_test)

# Avaliando a árvore de decisão
tree_accuracy = accuracy_score(y_test, tree_predictions)
tree_report = classification_report(y_test, tree_predictions)

print(f"Acurácia da Árvore de Decisão: {tree_accuracy}")
print(f"Relatório de Classificação da Árvore de Decisão:\n{tree_report}")
```

```python
# Treinando o modelo de Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Fazendo previsões com o Random Forest
rf_predictions = rf_model.predict(X_test)

# Avaliando o Random Forest
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_report = classification_report(y_test, rf_predictions)

print(f"Acurácia do Random Forest: {rf_accuracy}")
print(f"Relatório de Classificação do Random Forest:\n{rf_report}")
```
